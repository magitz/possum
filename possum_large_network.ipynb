```python
# Load the needed modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Normalization
from tensorflow.keras.initializers import RandomNormal, Zeros
```

```python
# Load the data from the file
df = pd.read_csv('https://raw.githubusercontent.com/magitz/possum/refs/heads/main/possum.csv')
df.head()
```

```python
import seaborn as sns
sns.pairplot(df, hue='sex')
```

```python
X=df[['hdlngth', 'skullw', 'skullw', 'totlngth', 'taill', 'footlgth', 'earconch', 'eye', 'chest', 'belly' ]]
y=df[['age']]
```

```python
# Instantiate a Sequential model, which allows us to build a neural network by stacking layers in a linear fashion.
model = Sequential()
model.add(Normalization())
model.add(Dense(units=512, activation='relu'))
model.add(Dense(units=512, activation='relu'))
model.add(Dense(units=256, activation='relu'))
model.add(Dense(units=128, activation='relu'))
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=32, activation='relu'))
model.add(Dense(units=1))
```

```python
# Compile the model (same as before)
model.compile(optimizer='adam', loss='mse', metrics=['mse'])  # Added 'mae' for accuracy
# Build the model with the input shape
# This initializes the weights of the model
model.build()  # (None, 2) for your input shape

model.summary()
```

```python
model.fit(X,y, validation_split=0.2, epochs=10 )
```
